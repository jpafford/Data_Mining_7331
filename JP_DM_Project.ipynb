```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore', DeprecationWarning)
%matplotlib inline
```

```python
excel_file = 'mathdownload2018.xlsx'
math2018=pd.read_excel(excel_file)
```

```python
#Checking the import and exploring the Data to see types

math2018.head()
```

```python
print (math2018.info())
```

```python
#Show summary statistics of numerical columns

math2018.describe()
```

```python
#Begining to Clean the data. The ultimage goal of this section is the label each row with an appropriate 
#label: state, county, public, charter, alternative, public district, charter district

math2018.loc[(math2018.charter_flag == 'Y') & (math2018.dass_flag != 'Y'), 'SchoolType'] = 'Charter'
math2018.loc[(math2018.charter_flag == 'Y') & (math2018.dass_flag == 'Y'), 'SchoolType'] = 'AltCharter'
math2018.loc[math2018.coe_flag == 'Y', 'SchoolType'] = 'County'
math2018.loc[(math2018.dass_flag == 'Y') & (math2018.charter_flag != 'Y'), 'SchoolType'] = 'AltPublic'
math2018.loc[(math2018.rtype == 'S') & (math2018.charter_flag != 'Y') & (math2018.dass_flag != 'Y'), 'SchoolType'] = 'Public'
math2018.loc[(math2018.rtype == 'X'), 'SchoolType'] = 'State'
math2018.loc[(math2018.rtype == 'D') & (math2018.coe_flag != 'Y'), 'SchoolType'] = 'District'
```

```python
#Sanity Check: Check new output to confirm SchoolType was created.

math2018.SchoolType.head()
```

```python
#Sanity Check: Check for nulls in the new SchoolType Column before deleting other columns, should have 7 unique values

math2018.SchoolType.describe()
```

```python
#Delete columns that are no longer necessary now that the school type is combined (rtype, charter_flag, coe_flag, dass_flag)
#Delete most of the ending columns that are all about alternate testing: profound disabilities (all the 'caa_...' columns)
#Delete reporting year (not needed since all data is 2018)

math2018 = math2018.drop(['rtype', 'charter_flag', 'coe_flag', 'dass_flag', 'caa_denom', 'caa_level1_num', 'caa_level1_pct', 'caa_level2_num', 'caa_level2_pct', 'caa_level3_num', 'caa_level3_pct', 'ReportingYear'], axis=1)
```

```python
#Sanity Check: Confirm delete processed appropriately - 12 less column

math2018.info()
```

```python
#We can also delete rows of data for campuses that are pairshare_method ('CA', 'DA', 'WA')
#there are 91 rows of data we can remove this way.
#Pairshare means that their scores were shared with another campus

math2018 = math2018.drop(math2018[(math2018.pairshare_method == 'CA') | (math2018.pairshare_method == 'DA') | (math2018.pairshare_method == 'WA')].index)
```

```python
#Sanity Check - should have all NaN in pairshare_method column

math2018.pairshare_method.describe()
```

```python
#Now that those lines have been deleted, we can eliminate the pairshare_method column

math2018 = math2018.drop(['pairshare_method'], axis=1)
```

```python
#Sanity Check: should have one less column

math2018.info()
```

```python
#What is the average change from 2017 to 2018 in distance from 'passing'?

math2018.change.mean()
```

```python
#Group School Type and Student Group and report average distance from 'passing' for 2018

math2018_grouped= math2018.groupby(by=['SchoolType', 'studentgroup'])
math2018_AvgDistance = math2018_grouped.currstatus.sum() / math2018_grouped.currstatus.count()
math2018_AvgDistance
```

```python
ax=math2018_AvgDistance.plot(kind='barh')
```

```python
#Chart of how many of each variable is represented (there are about 7 times as many public schools than charter schools)

pd.crosstab(math2018.SchoolType, math2018.studentgroup)
```

```python
#how the sub-pops are distributed across the School Types

pd.crosstab(math2018.SchoolType, math2018.studentgroup, normalize='columns')
```

```python
#Average Distance from 'Passing'

pd.crosstab(math2018.SchoolType, math2018.studentgroup, values = math2018.currstatus, aggfunc='mean').round(2)
```

```python
#Name it so I can use it.

math2018_AvgDist = pd.crosstab(math2018.SchoolType, math2018.studentgroup, values = math2018.currstatus, aggfunc='mean').round(2)
```

```python
math2018_AvgDist.plot(kind = 'bar', figsize=(15,15), stacked=False)
```
