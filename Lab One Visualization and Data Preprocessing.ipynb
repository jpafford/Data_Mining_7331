## Business Understanding
	
<p>&nbsp;&nbsp;&nbsp;&nbsp;This dataset contains information from schools within the California Department of Education. It specifically addresses school progress on standardized tests in Mathematics during the 2017-2018 school year. In part, the data will be used to support Local Education Agencies (LEA) in identifying strengths, weaknesses, and areas for improvement; assist in determining whether LEAs are eligible for assistance; and assist the SSPI in determining whether LEAs are eligible for more intensive state support/intervention. The school accountability system in California has 10 priorities. This dataset looks at one part of the 4th priority: Student Achievement. The dataset used in this analysis pertains specifically to the Mathematics curriculum and the progress made from 2017 school year to the 2018 school year.</p>
<p>&nbsp;&nbsp;&nbsp;&nbsp;From this dataset, we will look at how the state, each county, LEA, and campus increased or decreased their average distance above or below the passing standard score. This dataset also allows us to look at different programs (English Learner, Socioeconomic Disadvantaged, Students with Disabilities, Foster Youth, Homeless Youth) and race/ethnicities (Black/African American, American Indian or Alaska Native, Asian, Filipino, Hispanic, Pacific Islander, White, Multiple Races/Two or More). This project will address what differences in performance progress were seen between public and charter schools and its respective population subgroups, as described above. The outcome from this will provide further insights to California on the following:</p>

-	Is one subgroup out/underperforming another? 
-	How do districts compare performance wise? County Offices of Education? What about with subgroups?
-	Charter v. Traditional schools? What about with subgroups? 
-	Any impacts/trends by # of kids tested?
 
Data is important because:
-	It tells you how students performed on the California Assessment of Student Performance and Progress (CAASPP), california's state wide test.
-	It tells you what achievement/performance gaps exist - so policy can be made to address these issues.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore', DeprecationWarning)
%matplotlib inline

#Import data from data file and name it math2018

excel_file = 'mathdownload2018.xlsx'
math2018=pd.read_excel(excel_file)
math_df=pd.read_excel(excel_file)
```

## Data Quality

<p>&nbsp;&nbsp;&nbsp;&nbsp;Before we explore the data, one must be aware that the ‘All’ category total for any given campus/county/district data equals a number much smaller than if you were to add all the lines for similar campuses.  This is because a student may be in several different categories (i.e., a student may be White (WH), in foster care (FOS), be living in a low socio-economic status (SED) and be served by special education (SWD) – student will ‘count’ in 4 of the 16 categories listed). No edits were needed for this phenomenon, however, we need to be aware of this as we move forward.</p>

```python
#View basic info about the data file. There are many variables that have missing data.

math2018.info()
```

```python
#View initial findings of the "ALL" cateogry equaly all the other categories from the same campus.

math2018[['cds', 'studentgroup', 'currdenom']]
```

<p>&nbsp;&nbsp;&nbsp;&nbsp;Additionally, there were columns of data that we choose to eliminate from the data altogether.  There were 7 columns at the end of the dataset that give the data for the “California Alternate Assessment”.  This test is usually given to students served by special education who have profound disabilities and are unable to complete the same test as others.  There are approximately 1% of the total population who are included, and the data does not support our current research, therefore, these columns were deleted.  The column, "ReportingYear" was also removed because all data is from 2018.</p>

```python
#Delete most of the ending columns that are all about alternate testing: profound disabilities (all the 'caa_...' columns)
#Delete reporting year (not needed since all data is 2018)

math2018.drop(['caa_denom', 'caa_level1_num', 'caa_level1_pct', 'caa_level2_num', 'caa_level2_pct', 'caa_level3_num', 'caa_level3_pct', 'ReportingYear'], axis=1, inplace=True)
```

<p>&nbsp;&nbsp;&nbsp;&nbsp;Next, we found a column of data called “pairshare_method”.  If a campus does not include students of a testing grade level (i.e. PK campuses) or the campus only serves special populations (i.e., mental health hospital), the data is often paired with another campus in the district or charter.  If we kept this data in our dataset, these would be duplicates.  Therefore, the lines of data where there was a pair-share method were deleted and then the column was deleted because it was no longer needed.</p>


```python
#We can also delete rows of data for campuses that are pairshare_method ('CA', 'DA', 'WA')
#there are 91 rows of data we can remove this way.
#Pairshare means that their scores were shared with another campus

math2018.drop(math2018[(math2018.pairshare_method == 'CA') | (math2018.pairshare_method == 'DA') | (math2018.pairshare_method == 'WA')].index, inplace=True)

#Now that those lines have been deleted, we can eliminate the pairshare_method 

math2018.drop(['pairshare_method'], axis=1, inplace=True)
```

<p>&nbsp;&nbsp;&nbsp;&nbsp;There are many missing values, however, they are not mistakes. The values that are missing appear when a population has 10 or fewer students represented. When this occurs, the "currstatus" and "change" columns are not calculated for the purpose of this data file and state accountability ratings (to protect data privacy). We chose to delete these lines from our dataset because the data we need to complete our analysis is not included for these lines. The above steps of removing data led to dropping our dataset from 148,932 entries to 97,396.</p>

```python
#Delete any row of data where there is no Currstatus data - meaning that there were less than 10 students that make up that group

math2018.dropna(subset=["currstatus"], inplace=True)
```

```python
#Perform a 'sanity check' to ensure we took out the data that needed to be reomved

math2018.info()
```

## Explore Joint Attributes
<p>&nbsp;&nbsp;&nbsp;&nbsp;In order to get a better look at the differentiation between different school types, we created a new attribute, "SchoolType". This new attribute was needed because the different types of schools were not noted in one single column. After doing so, we used groupby to create a group-wise average. Once the new column, called "SchoolType" was created, the former columns were deleted.</p>

```python
#Create a new column, SchoolType; label each row with an appropriate label: state, county, district, public, charter, alternative public, and alternative charter

math2018.loc[(math2018.charter_flag == 'Y') & (math2018.dass_flag != 'Y'), 'SchoolType'] = 'Charter'
math2018.loc[(math2018.charter_flag == 'Y') & (math2018.dass_flag == 'Y'), 'SchoolType'] = 'AltCharter'
math2018.loc[math2018.coe_flag == 'Y', 'SchoolType'] = 'County'
math2018.loc[(math2018.dass_flag == 'Y') & (math2018.charter_flag != 'Y'), 'SchoolType'] = 'AltPublic'
math2018.loc[(math2018.rtype == 'S') & (math2018.charter_flag != 'Y') & (math2018.dass_flag != 'Y'), 'SchoolType'] = 'Public'
math2018.loc[(math2018.rtype == 'X'), 'SchoolType'] = 'State'
math2018.loc[(math2018.rtype == 'D') & (math2018.coe_flag != 'Y'), 'SchoolType'] = 'District'

#Delete columns that are no longer necessary now that the school type is combined (rtype, charter_flag, coe_flag, dass_flag)

math2018.drop(['rtype', 'charter_flag', 'coe_flag', 'dass_flag'], axis=1, inplace=True)
```

<p>&nbsp;&nbsp;&nbsp;&nbsp;As we explored the data further, we found the need to drop a few more aspects of our data set. One final step in the initial cleaning process was to remove the State, County and District lines of data. Leaving these lines of data in the file, leads to duplicates. Every school belongs to 3 larger organizations - it's respective district, county, and state.</p>

```python
math2018.drop(math2018[(math2018.SchoolType=='State') | (math2018.SchoolType=='County') | (math2018.SchoolType=='District')].index, inplace=True)
```

<p>&nbsp;&nbsp;&nbsp;&nbsp;Once we removed the additional lines of data, we were easily able to view the grouped table to begin the analysis.</p>

```python
#Group School Type and Student Group and report average distance from 'passing' for 2018

math2018_grouped= math2018.groupby(by=['SchoolType', 'studentgroup'])
math2018_AvgDistance = math2018_grouped.currstatus.sum() / math2018_grouped.currstatus.count()
math2018_AvgDistance
```

<p>&nbsp;&nbsp;&nbsp;&nbsp;When we looked at these data, we found some significant differences in the average distance from passing between the different school types and student groups. Additionally, we created a cross-tabulation chart to enable us to visualize the information in barplot form.  Seeing this visually, enabled us to see the stark differences in the average distance from passing in each school type.<p>

```python
#Average Distance from 'Passing'

pd.crosstab(math2018.SchoolType, math2018.studentgroup, values = math2018.currstatus, aggfunc='mean').round(2)
```

## Visualize Attributes
<p>&nbsp;&nbsp;&nbsp;&nbsp;We will visualize the percentage of missing values for every feature and highlight the features with a threshold of less than 40% missing.  This will aid in helping us understand the data set more and focus in on the features that is relevant to insights and conclusion.  Judging by the first bar chart, there are a lot of features that have missing data above the 40% threshold.  The second chart subsets all the feature and shows only the features that meet the threshold limit.</p>

```python
#calcuate the missing percentage for each of the features
missing = math_df.isnull().sum()/len(math_df)*100

dx=missing.plot(kind='barh', figsize=(18,20))
plt.title('Percentage of Missing Values for Every Feature')
plt.xlabel('Percentages')
```

```python
# saving column names in a list called variable
variables = math_df.columns
variable = []

# setting threshold as 40% to impute features above the threshold
for i in range (0,32):
    if missing[i]<=40: 
        variable.append(variables[i])

math2018_sub = math_df[variable]
missing_perc = math2018_sub.isnull().sum()/len(math2018)*100
        
ex=missing_perc.plot(kind='barh', figsize=(18,20))
plt.title('Features with less than 40% missing')
plt.xlabel('Percentages')
```

```python
#perform univariate feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

def featureselection(table):
    array = table.values #change pandas dataframe to numpy array
    array[np.isnan(array)] = 0 #impute all NaN to 0
    _input = array[:,0:16] #parse out features to be filtered
    _target = array[:,16] #parse out response variable

    #feature selecton using SelectKBest with f_regression as scoring function
    test = SelectKBest(score_func=f_regression, k=4)
    fit = test.fit(_input, _target)
    features = fit.transform(_input)

    #match scores to feature names and display
    index=list(table)[:-1]
    np.set_printoptions(precision=3) 
    values=fit.scores_
    featurescores = pd.DataFrame(values, index=index, columns=["scores"])
    featurescores.sort_values(by=['scores'], inplace=True, ascending=False)
    print(featurescores[:5])
    
featureselection(newtable)
```
